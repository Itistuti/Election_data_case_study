```python
joined_df=election_df.merge(district_df, on='district_id',how='inner').merge(province_df, on='province_id')
joined_df.groupby('province_name')['votes'].sum().sort_values(ascending=False).head()
```

#### 8. For all constituencies, find the `winner` and `first runner-up` and assign that output to variable `top_2`

```python
# Code here
top_2 = election_df.sort_values(
    ["const_name", "votes"],
    ascending=[True, False]
).groupby("const_name").head(2)

top_2
```

#### 9. Compute the `margin` between `winner` and `first runner-up` for each constituency. 

```python
# Code here
margin_df = top_2.groupby("const_name")["votes"].diff().abs()
margin_df = top_2.assign(margin=margin_df).dropna()[["const_name", "margin"]]
margin_df
```

#### 10. Identifies the top 10 most competitive constituency (lowest margins).

```python
# Code here
margin_df.sort_values("margin").head(10)
```
